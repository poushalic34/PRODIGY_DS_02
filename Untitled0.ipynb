```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = pd.read_csv("/content/drive/MyDrive/Amazon-Products.csv")
```

```python
a.head(5)
```

```python
a.drop(["Unnamed: 0","image","link"], axis=1, inplace=True)
```

```python
a.head(5)
```

```python
a.info()
```

```python
a.isnull().sum()
```

```python
i = a[a.actual_price.isnull()].index
a.drop(i, axis=0, inplace=True)
```

```python
a.isnull().sum()
```

```python
a[a.discount_price.isnull()]
```

```python
rows = a[a.discount_price.isnull()].index

for i in rows:
    a.loc[i]["discount_price"] = a.loc[i]["actual_price"]
```

```python
a.isnull().sum()
```

```python
a[a.ratings.isnull() & a.no_of_ratings.isnull()]
```

```python
a.ratings.fillna(0, inplace=True)
a.no_of_ratings.fillna(0, inplace=True)
```

```python
a.isnull().sum()
```

```python
a.duplicated().sum()
```

```python
a[a.duplicated()]
```

```python
a.drop_duplicates(inplace=True)
```

```python
a.replace([",","₹"],"", regex= True, inplace= True)
```

```python
a.head(2)
```

```python
a.ratings.unique()
```

```python
a["ratings"].replace(["Get","FREE"],"0",inplace= True)
```

```python
i2 = a[a.ratings == '2.99' ].index
a.drop(i2,axis=0, inplace = True)
```

```python
a.drop((a[a.ratings == '68.99' ].index), axis=0, inplace = True)
a.drop((a[a.ratings == '65' ].index), axis=0, inplace = True)
a.drop((a[a.ratings == '70' ].index), axis=0, inplace = True)
a.drop((a[a.ratings == '100' ].index), axis=0, inplace = True)
a.drop((a[a.ratings == '99' ].index), axis=0, inplace = True)
```

```python
a.ratings.unique()
```

```python
a.no_of_ratings.unique()
```

```python
a["no_of_ratings"].replace(["Only 2 left in stock.",'Only 1 left in stock.','FREE Delivery by Amazon',
                                 'Usually dispatched in 2 to 3 days.','Usually dispatched in 5 to 6 days.',
                                 'Usually dispatched in 4 to 5 days.','Usually dispatched in 6 to 7 days.',
                                 'Usually dispatched in 7 to 8 days.','Only 3 left in stock.','Usually dispatched in 3 to 4 weeks.',
                                 'Usually dispatched in 4 to 5 weeks.','Only 5 left in stock.','Only 4 left in stock.',
                                 'Usually dispatched in 3 to 5 days.','Usually dispatched in 11 to 12 days.',
                                 'Usually dispatched in 9 to 10 days.','Usually dispatched in 2 to 3 weeks.',
                                 'Usually dispatched in 8 to 9 days.','Usually dispatched in 1 to 2 months.',
                                 'This item will be released on August 14 2023.']
                                 , "0",inplace= True)
```

```python
a[["ratings","no_of_ratings","discount_price","actual_price"]] = a[
    ["ratings","no_of_ratings","discount_price","actual_price"]].astype('float')
```

```python
a.info()
```

```python
a.head()
```

```python
a["Discount"] = a["actual_price"] - a["discount_price"]
```

```python
a["Discount_Percentage(%)"] = round((a["Discount"]/a["actual_price"])*100,2)
```

```python
a.sample(5)
```

```python
a["Manufacture"] = a.name.str.split(" ").str[0]
```

```python
a.head()
```

```python
a.Manufacture.value_counts(ascending=False).head(15)
```

```python
a.main_category.value_counts(ascending=False).head(15)
```

```python
plt.figure(figsize=(8,4))
(a.main_category.value_counts(ascending=False).head(10)).plot(kind="bar", xlabel="Main Category",ylabel="Total Products")
plt.title("\n Top 10 Product Categories \n")
plt.show()
```

```python
a.sub_category.value_counts(ascending=False).head(10)
```

```python
toprating.main_category.value_counts().head(10).plot(kind="bar",figsize=(10,4), xlabel="Category",ylabel="Count")
plt.title("\n Top Rated Category (Top 10) \n")
plt.show()
```

```python
toprating.sub_category.value_counts().head(10).plot(kind="bar",figsize=(10,4), xlabel="Sub-Category", ylabel="Count")
plt.title("\n Top Rated Sub-Category (Top 10) \n")
plt.show()
```

```python
df = a[a.no_of_ratings != 0.0 ] #Excluding '0' values
df.groupby("main_category")['no_of_ratings'].mean().sort_values(ascending=False).plot(kind='bar',figsize=(11,4))
plt.title("Total No. of Ratings - Main Category \n")
plt.ylabel=("Total Ratings")
plt.show()
```

```python
df.groupby("Manufacture")['no_of_ratings'].mean().sort_values(ascending=False).head(15).plot(kind='bar',figsize=(11,4))
plt.title("Total No. of Ratings for Manufacturer \n")
plt.ylabel=("Total Ratings")
plt.show()
```
